```python
import unittest
import nose.tools as nt
import numpy as np
import tools

class test_tools():

    def setUp(self):
        # Create a random array
        self.n = 4
        self.rand_array = np.random.normal(size=(self.n,self.n))

    def tearDown(self):
        pass

    def test_square(self):
        print(self.rand_array)

        # Test the square function to see if it
        # returns an array of the right size
        output_arr = tools.square(self.rand_array)
        nt.assert_equal(output_arr.shape, (self.n, self.n))

        # Test the square function to make sure that squaring
        # a matrix containing only 1s and 0s returns the same thing
        test_arr = np.diag(np.ones(self.n))
        output_arr = tools.square(test_arr)
        for i in range(self.n):
            for j in range(self.n):
                nt.assert_equal(output_arr[i,j], test_arr[i,j])

        # Test the square function for a known input/output
        test_arr = np.array([1, 2])
        output_arr = tools.square(test_arr)
        nt.assert_equal(output_arr[0], 1)
        nt.assert_equal(output_arr[1], 4)


    def test_get_pi(self):
        # See if the get_pi function really returns pi
        alleged_pi = tools.get_pi()
        nt.assert_almost_equal(alleged_pi, 3.141592653589)

    def test_picky(self):
        # Make sure that tools.picky raises an error if the
        # wrong type is inputted
        nt.assert_raises(TypeError, tools.picky, 'hey')
        
    def checkPeak(self):
        #MAKE dummy function to test peak, over here we ake gaussian centered at 0
        # Plot between -10 and 10 with .001 steps.
        x_axis = np.arange(-20, 20, 0.01)

        # Calculating mean and standard deviation
        mean = statistics.mean(x_axis)
        sd = statistics.stdev(x_axis)
        nt.assert_equal(find_peak(norm.pdf(x_axis, mean, sd)), 0)
        

    def checkPosteriorCalculation(self):
        #eg find life at 1,2,3 at multiply it seperately, given by post
        x = np.linspace(0,2,1000)
        dx = x[1]-x[0]
        y1 =  life([1], x)
        y2 =  life([2], x)
        y3 = life([3], x)
        post = y1*y2*y3
        norm_post = post/(np.sum(post*dx))
        #call our function on similar values and subtrat from manually multiplied post, functions should get 0 
        yy = norm_post - posterior_2([1,2,3], [], x)
        nt.assert_equal(np.sum(yy), 0)
        
    def RateParameter1(self):
        #should return 1.2599210498948732
        nt.assert_equal(step_fn(1, 1.1), 1.2599210498948732)

    def RateParameter2(self):
        nt.assert_equal(step_fn(1, 0.9), 0)

    def RateParameter3(self):
        nt.assert_equal(linear_fn(1, 0.9,10), 9)
        
    def RateParameter3(self):
        nt.assert_equal(linear_fn(1, 1.1,10), 1.2599210498948732)
    
    def RateParameter4(self):
        nt.assert_equal(exp_fn(1, 0.5,10), 2.5)
        
    def RateParameter5(self):
        nt.assert_equal(exp_fn(1, 1,10), 1.2599210498948732)
```
