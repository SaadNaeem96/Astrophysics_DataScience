We first start my simulating a sequence of Poisson events. Let us define first a function returning a Poisson waiting time associated to a rate $r$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from random import *
def wait_time(r):
    return -1/float(r)*np.log(random())
```

Now we are going to define a simple function simulating Poisson events until we reach a time window of size $S$, and returning the number of events

```python
def run_Poisson_window(S,r):
    t=0
    n=0
    while (t<S):
        n=n+1
        t=t+wait_time(r)
    
    return n-1
```

Now we are going many instances of this problem, accumulate stats to get the distribution $P(N)$. We are going to compare to the Poisson Distribution.

```python
list=[]
S=30
r=0.1
for i in range(100000):
    n=run_Poisson_window(S,r)
    list.append(n)
#Comparison to Poisson Distribution

def Poisson(n,average):
    return average**n*np.exp(-average)/np.math.factorial(n)

Poisson_distrib=[]
for i in range(20):
    Poisson_distrib.append(Poisson(i,S*r))
```

```python
array=np.histogram(list,density=True,bins=range(20))
plt.plot(array[0],marker='.')
plt.plot(range(20),Poisson_distrib)
```



We illustrate now the Gillespie algorithm on the simple production/degradation process as studied in class this morning.

A protein is produced with a rate $\rho$ and degraded with rate $\delta$.
We compute propensities, compute next time $\tau$ and next reaction (there are only two of them).
Then we update our array of results  `result`. The array `result_determ` contains the deterministic equation for comparison

```python
npoints=50000
size=2
rho=0.5
delta=0.1
result=np.zeros((size,npoints),dtype=float)
result_determ=np.zeros((size,npoints),dtype=float)
seed(40)
result[1,0]=0

#those three variables are used to properly sample protein numbers at equal times to get their distribution
sample_step=20
current_time=2000
sampling_proteins=[] 
for i in range(npoints-1):
    nprot=result[1,i] #take number of proteins
    propens_production=rho
    propens_degradation=nprot*delta
    a_0=propens_production+propens_degradation
    print(a_0)
    r1=random()
    r2=random()
    tau=-1/float(a_0)*np.log(r1)
    result[0,i+1]=result[0,i]+tau
    
    if (propens_production>r2*a_0):
        result[1,i+1]=result[1,i]+1
    else:
        result[1,i+1]=result[1,i]-1
    result_determ[0,i+1]=result[0,i+1]
    result_determ[1,i+1]=rho/float(delta)*(1-np.exp(-delta*result_determ[0,i+1]))
    if (result[0,i+1]>current_time):
        # sampling of protein numbers at equal times
        sampling_proteins.append(result[1,i+1])
        current_time=current_time+sample_step
```

Now we plot  everything

```python
plt.plot(result[0,:],result[1,:])
plt.plot(result_determ[0,:],result_determ[1,:])
```

Comparison of protein distributions with Poisson distribution

```python
Poisson_distrib=[]
for i in range(30):
    Poisson_distrib.append(Poisson(i,rho/delta))
array=np.histogram(sampling_proteins,density=True,bins=range(30))
plt.plot(array[0],marker='.')
plt.plot(range(30),Poisson_distrib,marker='.')
```
